```python
# import the libraries
import requests
import pandas as pd
```

```python
#store the link
url = 'https://earthquake.usgs.gov/fdsnws/event/1/application.json'

#check to see if we get a response from the API
r = requests.get(url)
r
months = ["01","02","03","04","05","06","07","08","09","10","11","12"]
data = []
for y in range(1920,2023):
    for m in months:
        x =requests.get(f'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={str(y)}-{m}-01&endtime={str(y)}-{m}-31')
        data.append(x)
        
      
```

```python
import pickle
```

```python
pickle_out = open("list.pickle","wb")
pickle.dump(data, pickle_out)
pickle_out.close()

# https://pythonprogramming.net/python-pickle-module-save-objects-serialization/#:~:text=First%2C%20import%20pickle%20to%20use,into%20opened%20file%2C%20then%20close.&text=Use%20pickle.,-load()%20to
```
